```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
```

```python
df
```

```python
# We doot need the name column here
df.drop("name",axis=1, inplace=True)
```

```python
# Finding the number of NullValues 
df.isna().sum()
```

```python
# Since we have not done any outlier treatemnt so we have to replae the ull Values in HorePower Column with median value 
df["horsepower"] = df["horsepower"].fillna(df.horsepower.median())

# determine the value of horsepower Median 
df.horsepower.median()
```

```python
df["horsepower"]
```

```python
# Again Finding the number of NullValues 
df.isna().sum()
```

```python
# Determinng the data Type of the columns in the dataset
df.info()
# We need to change the datatype of the object type to numerical values 
```

```python
# # So we need to change the values in the origin column forom object type to integer type 
# # Lets first determine the diffrent ypes of categiors of values in the origin columns
df["origin"].value_counts()
```

```python
df["origin"] = df["origin"].map({"usa":1,"japan":2,"europe":3})
# # # Converting the datatype of the origin columns into int64
df["origin"] = df["origin"].astype(int)

```

```python
df["origin"].value_counts()
```

```python
df.info() # Now we can see all the features have the datatype intgere or float 
```

```python
# You need to seperate the X and y 
# Target variable (y)
y = df['mpg']

# Feature variables (X) - all columns except 'mpg'
X = df.drop('mpg', axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Assuming X and y are already defined as features and target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# This will split 80% of the data into training and 20% into testing
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize the Linear Regression model
regression_model = LinearRegression()
```

```python
# Fit the model using the training data
regression_model.fit(X_train, y_train)
```

```python
for index, column_name in enumerate(X_train.columns):
    print(column_name,regression_model.coef_[index])
```

```python
# This will give you the R-squared score for your linear regression model, specifically for the test set

from sklearn.metrics import r2_score

# Predicting on the test data
y_pred_linear = regression_model.predict(X_test)

# Calculate the R-squared score for the test data
r2_linear = r2_score(y_test, y_pred_linear)

print(f"R-squared score for the test set: {r2_linear}")
```

```python
from sklearn.linear_model import Ridge
```

```python
# Initialize the Ridge Regression model
ridge_regression_model = Ridge(alpha=1.0) 
```

```python
# Fit the Ridge model using the training data
ridge_regression_model.fit(X_train, y_train)
```

```python
for index, column_name in enumerate(X_train.columns):
    print(column_name,ridge_regression_model.coef_[index])
# So here we can see there is not much change in the coefficients in regression model and ridge_regression model
# This means that there is no much learning by the model after L2 regualrization 
```

```python
# Predicting on the test data
y_pred_ridge = ridge_regression_model.predict(X_test)

# Calculate the R-squared score for the test data
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f"R-squared score for the test set: {r2_ridge}")
# We dont see much variation in R2 score of ridge regression than Linear regression 
```

```python
from sklearn.linear_model import Lasso
```

```python
# Initialize the Lasso Regression model
lasso_regression_model = Lasso(alpha=0.5)  # Adjust the alpha value if needed
```

```python
# Fit the Lasso model using the training data
lasso_regression_model.fit(X_train, y_train)
```

```python
# Coefficients for Lasso Regression
print("Lasso Regression Coefficients:")
for index, column_name in enumerate(X_train.columns):
    print(column_name, lasso_regression_model.coef_[index])
```

```python
# Predicting on the test data using Lasso Regression
y_pred_lasso = lasso_regression_model.predict(X_test)
```

```python
# Calculate the R-squared score for the test data using Lasso Regression
r2_lasso = r2_score(y_test, y_pred_lasso)

# Print the R-squared score
print(f"R-squared score for the test set (Lasso Regression): {r2_lasso}")

# Now we an see that the r2scoe with lasso regularization has redued from 83 to 82 as comaprred to Linear regresson r2 score 
# Tha Means lassso has eliminated soe of the unimportant features 
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import r2_score

# Initialize the ElasticNetCV model
elastic_net_cv_model = ElasticNetCV(cv=5)  # You can adjust the alpha values and cv folds if needed

# Fit the ElasticNetCV model using the training data
elastic_net_cv_model.fit(X_train, y_train)

# Get the best alpha and l1_ratio found by ElasticNetCV
best_alpha = elastic_net_cv_model.alpha_
best_l1_ratio = elastic_net_cv_model.l1_ratio_
print(f"Best alpha value found: {best_alpha}")
print(f"Best l1_ratio value found: {best_l1_ratio}")

# Get the coefficients of the ElasticNetCV model
coefficients = elastic_net_cv_model.coef_
print("ElasticNetCV Regression Coefficients:")
for index, column_name in enumerate(X_train.columns):
    print(f"{column_name}: {coefficients[index]}")

# Predicting on the test data using ElasticNetCV
y_pred_elastic_net_cv = elastic_net_cv_model.predict(X_test)

# Calculate the R-squared score for the test data using ElasticNetCV
r2_elastic_net_cv = r2_score(y_test, y_pred_elastic_net_cv)

# Print the R-squared score
print(f"R-squared score for the test set (ElasticNetCV Regression): {r2_elastic_net_cv}")

# Optionally, you can also get the list of parameters used in ElasticNetCV
print("ElasticNetCV parameters:")
print(f"Alpha values tested: {elastic_net_cv_model.alphas_}")
```
